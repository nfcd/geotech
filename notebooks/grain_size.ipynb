## Laboratory on geology, part B

A soil type from the shoreline in Møllebukta that was analyzed a couple of years ago had the following grain-size distribution: OBS: Table in the lab handout.

1. Give the soil type an appropriate name.
2. Calculate $C_u$. Is the soil well-graded or poorly graded?
3. Evaluate the soil’s frost susceptibility.

```python
# sieve (grain) size in mm
gz = [8, 4, 2, 1, 0.5, 0.25, 0.125, 0.063]

# soil mass in sieve in grams
j = np.array([12.8, 43.3, 66, 54.5, 59.9, 68.1, 40.2, 5.6])
total = np.sum(j) + 0.1  # add 0.1 g in pan
# compute percent passing
p = 100 * (1 - np.cumsum(j) / total)

# find the percentages of gravel, sand, silt, and clay
gravel = 100 - np.interp(2.0, gz[::-1], p[::-1])
sand = 100 - gravel - np.interp(0.063, gz[::-1], p[::-1]) 
silt = 100 - gravel - sand 
print(f"Gravel: {gravel:.2f} %, Sand: {sand:.2f} %, Silt: {silt:.2f} %")

# compute coefficient of uniformity (Cu) 
# D60 is the grain size at 60% passing
D60 = np.interp(60, p[::-1], gz[::-1])
# D10 is the grain size at 10% passing
D10 = np.interp(10, p[::-1], gz[::-1])
# compute coefficient of uniformity
Cu = D60 / D10
print(f"D60 = {D60:.3f} mm, D10 = {D10:.3f} mm, Cu = {Cu:.2f}")

# Create a figure and axis 
fig, ax = plt.subplots(figsize=(13.5, 6.5))

# plot the grain size chart in Norwegian
grain_size_chart(fig, ax, English=False)

# plot soil
ax.plot(gz, p, '*-', label='Soil at Møllebukta', color='black')

ax.legend()

plt.show()
```

## Exercise 1.1 of Aarhaug (1984)

Two grain-size distribution analyses have been carried out, and the results are given in Table 1.4. Plot the grain-size distribution curves for soil types A and B and determine the percentage content of gravel, sand, silt, and clay. Determine the grain-size distribution coefficient $C_u$, and determine the frost susceptibility class $T$ for the soil types. What are the names of the soil types?

```python
# sieve (grain) size in mm
gz = [60, 10, 5, 2, 1, 0.6, 0.2, 0.1, 0.06, 0.02, 0.01, 0.006, 0.002]

# soil A:

# soil mass in sieve in grams
j = np.array([0, 75, 35, 55, 50, 35, 65, 35, 25, 45, 25, 15, 20])
total = np.sum(j) + 20  # add 20 g in pan
# compute percent passing
pa = 100 * (1 - np.cumsum(j) / total)

# find the percentages of gravel, sand, silt, and clay
gravel = 100 - np.interp(2.0, gz[::-1], pa[::-1])
sand = 100 - gravel - np.interp(0.063, gz[::-1], pa[::-1]) 
silt = 100 - gravel - sand - np.interp(0.002, gz[::-1], pa[::-1])
clay = 100 - gravel - sand - silt
print("Soil A")
print(f"Gravel: {gravel:.2f} %, Sand: {sand:.2f} %, Silt: {silt:.2f} %, Clay: {clay:.2f} %")

# compute coefficient of uniformity (Cu)
# D60 is the grain size at 60% passing
D60 = np.interp(60, pa[::-1], gz[::-1])
# D10 is the grain size at 10% passing
D10 = np.interp(10, pa[::-1], gz[::-1])
# compute coefficient of uniformity
Cu = D60 / D10
print(f"D60 = {D60:.3f} mm, D10 = {D10:.3f} mm, Cu = {Cu:.2f}")

# soil B:

# soil mass in sieve in grams
j = np.array([0, 0, 0, 0, 0, 10, 110, 130, 90, 120, 35, 5, 0])
total_b = np.sum(j) + 0  # add 0 g in pan
# compute percent passing
pb = 100 * (1 - np.cumsum(j) / total_b)

# find the percentages of gravel, sand, silt, and clay
gravel = 100 - np.interp(2.0, gz[::-1], pb[::-1])
sand = 100 - gravel - np.interp(0.063, gz[::-1], pb[::-1]) 
silt = 100 - gravel - sand - np.interp(0.002, gz[::-1], pb[::-1])
clay = 100 - gravel - sand - silt
print("")
print("Soil B")
print(f"Gravel: {gravel:.2f} %, Sand: {sand:.2f} %, Silt: {silt:.2f} %, Clay: {clay:.2f} %")

# compute coefficient of uniformity (Cu)
# D60 is the grain size at 60% passing
D60 = np.interp(60, pb[::-1], gz[::-1])
# D10 is the grain size at 10% passing
D10 = np.interp(10, pb[::-1], gz[::-1])
# compute coefficient of uniformity
Cu = D60 / D10
print(f"D60 = {D60:.3f} mm, D10 = {D10:.3f} mm, Cu = {Cu:.2f}")

# Create a figure and axis 
fig, ax = plt.subplots(figsize=(13.5, 6.5))

# plot the grain size chart in Norwegian
grain_size_chart(fig, ax, English=False)

# plot soil A
ax.plot(gz, pa, '*-', label='Soil A')

# plot soil B
ax.plot(gz, pb, '*-', label='Soil B')

ax.legend()

plt.show()
```
