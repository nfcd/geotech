# Surface area:

### Example 1.3 of Aarhaug (1984):

We consider 1 $\mathrm{cm^3}$ of solid sand that has an average grain size, simplified to 1 $\mathrm{mm^3}$.



The total surface area is:

```python
# Number of sand grains in 1 cm
num_grains_sand = 10 * 10 * 10 
print(f"Number of sand grains in 1 cm: {num_grains_sand}")

# surface area of one sand grain 
surface_area_grain_sand = 6 * (1 ** 2) # in mm
print(f"Surface area of one sand grain: {surface_area_grain_sand} mm")

# Total surface area of sand grains in 1 cm
total_surface_area_sand = (surface_area_grain_sand * num_grains_sand) / 1e6 # in m
print(f"Total surface area of sand grains in 1 cm: {total_surface_area_sand} m")  
```

### Example 1.4 of Aarhaug (1984):

We consider 1 $\mathrm{cm^3}$ of clay with mineral grain size of length x width x thickness = $1\mathrm{\mu m}\cdot1\mathrm{\mu m}\cdot0.1\mathrm{\mu m}$. 



The total surface area is:

```python
# Number of clay grains in 1 cm
num_grains_clay = 1e3 * 1e3 * 1e4 * 1e3 
print(f"Number of clay grains in 1 cm: {num_grains_clay:.1e}")

# surface area of one clay grain
surface_area_grain_clay = 2 * 1e-3 * 1e-3 + 4 * 1e-3 * 1e-4 # in mm
print(f"Surface area of one clay grain: {surface_area_grain_clay} mm")

# Total surface area of clay grains in 1 cm
total_surface_area_clay = (surface_area_grain_clay * num_grains_clay) / 1e6 # in m
print(f"Total surface area of clay grains in 1 cm: {total_surface_area_clay} m")
```

The ratio of total surface area between clay and sand is:

```python
# Ratio of surface areas
ratio_surface_areas = total_surface_area_clay / total_surface_area_sand
print(f"Ratio of total surface area of clay to sand: {ratio_surface_areas}")
```
